```python
import torch

def replace_path_in_obj(obj, old_path, new_path):
    if isinstance(obj, dict):
        for key, value in obj.items():
            if isinstance(value, (dict, list)):
                replace_path_in_obj(value, old_path, new_path)
            elif isinstance(value, str):
                obj[key] = value.replace(old_path, new_path)
    elif isinstance(obj, list):
        for i in range(len(obj)):
            if isinstance(obj[i], (dict, list)):
                replace_path_in_obj(obj[i], old_path, new_path)
            elif isinstance(obj[i], str):
                obj[i] = obj[i].replace(old_path, new_path)

# 加载.pth文件
net = torch.load('data\\samples\\Syria_samples\\split_size6\\train.pth')

# 检查并修改字典中的路径
replace_path_in_obj(net, '/ssd/hk/upload_samples', 'data/samples')

# 保存修改后的字典到.pth文件
torch.save(net, 'data\\samples\\Syria_samples\\split_size6\\train.pth')

net
```
